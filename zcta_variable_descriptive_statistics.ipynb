```python
import matplotlib 
import matplotlib.pyplot as plt
import pandas as pd

matplotlib.style.use('ggplot')
```

```python
# Load the data
tjhd_df = pd.read_csv('tjhd_zcta_variables.csv')
print(tjhd_df.columns)
```

```python
# Define the list of variables to use in the analysis
variables = ['Age_Cat0_Pct', 'Age_Cat1_Pct', 'Age_Cat2_Pct', 'Age_Cat3_Pct', 'Age_Cat4_Pct',
             'Inc_Cat0_Pct', 'Inc_Cat1_Pct', 'Inc_Cat2_Pct',
             'Median_HH_Inc',
             'Bachelor_Plus_Pct',
             'White_Pct', 'Black_Pct', 'Asian_Pct', 'Hispanic_Pct',
             'Owner_Occupied_Pct',
             'Urban_Pct',
             'Population_Density']
```

```python
# Pairs Plot
pd.scatter_matrix(tjhd_df[variables], figsize=(50, 50))
plt.show()
plt.close()
```

```python
# Grouped box plots
groups = [('Age', ['Age_Cat0_Pct', 'Age_Cat1_Pct', 'Age_Cat2_Pct', 'Age_Cat3_Pct', 'Age_Cat4_Pct']),
          ('Household Income 1', ['Inc_Cat0_Pct', 'Inc_Cat1_Pct', 'Inc_Cat2_Pct']),
          ('Household Income 2', ['Median_HH_Inc']),
          ('Educational Attainment', ['Bachelor_Plus_Pct']),
          ('Race', ['White_Pct', 'Black_Pct', 'Asian_Pct', 'Hispanic_Pct']),
          ('Owner Occupied Household', ['Owner_Occupied_Pct']),
          ('Urban vs. Rural', ['Urban_Pct']),
          ('Population Density', ['Population_Density'])]

for grp_name, grp in groups:
    plt.figure(figsize=(12, 8))
    tjhd_df.boxplot(column=grp)
    plt.title(grp_name)
    plt.show()
    plt.close()
```
